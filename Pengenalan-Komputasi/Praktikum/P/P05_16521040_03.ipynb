```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('P05-03.csv')

print("2. Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam diatas 300")
data = df.loc[df['price'] > 300]
# print(data)
jumlah = data['neighbourhood'].value_counts()
# print(jumlah)
jumlah.plot(kind = "pie")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('P05-03.csv')

print("3. Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neigh. Kallang")
data = df.loc[df['neighbourhood'] == 'Kallang']
data.plot(kind="scatter", x="longitude", y="latitude")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('P05-03.csv')

print("4. Bubble plot dengan x : long, y : lat, dan banyaknya review sebagai ukuran dari rumah")
print("dengan banyak hari ketersediaan rumah dalam satu tahun diatas 100")
data = df.loc[df['availability_365'] > 100]
data.plot(kind="scatter",x="longitude", y="latitude",sizes=data['number_of_reviews'], color="red")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('P05-03.csv')

print("5. Stacked bar chart dari rumah yang disewakan oleh Richards dengan x : neigh. dan y : jenis penyediaan ruang")
data = df.loc[df['host_name'] == 'Richards']
# print(data)
data.groupby(['neighbourhood','room_type']).size().unstack().plot(kind="bar",stacked=True)
# Pembuktian bahwa tidak ada data lain untuk room_type yang disewakan oleh Richards
print(data.loc[df['room_type'] != 'Entire home/apt'])
print(data.loc[df['room_type'] == 'Private room'])

# print(data)
# print(data.loc[df['room_type'] == 'Entire home/apt'])
if len(data) == len(data.loc[df['room_type'] == 'Entire home/apt']):
    print("Data sama")
```
